## Wordcloud for Lincoln's 2nd Inaugural Address

Visualizations inspired by Andreas Mueller's stylistic word clouds (https://github.com/amueller/word_cloud) and completed with many thanks to Sheena Lee Villaneuva (https://github.com/sleevillanueva)

```python
import numpy as np
import csv
import random
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
from scipy.misc import imread
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
text = open('../Viz/lincoln.txt').read()
```

```python
linc_mask = imread('../Viz/linc_sten.png')
```

```python
def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(10, 50)
```

```python
wc = WordCloud(background_color='white', mask=linc_mask, stopwords=STOPWORDS)
```

```python
wc.generate(text)
```

```python
wc.recolor(color_func=grey_color_func, random_state=3)
```

```python
plt.imshow(wc)
plt.axis('off')
plt.figure()
```

```python
wc.to_file('../Viz/linc_wc.png')
```

## Wordcloud for Oscar Wilde's corpus

Visualizations inspired by Andreas Mueller's stylistic word clouds (https://github.com/amueller/word_cloud) and completed with many thanks to Sheena Lee Villaneuva (https://github.com/sleevillanueva)

```python
import numpy as np
import csv
import random
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from scipy.misc import imread
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
text = open('../Viz/_CLEAN_Wilde.txt').read()
```

```python
image = imread('../Viz/green.png')
```

```python
grn = ImageColorGenerator(image)
```

```python
wilde_mask = imread('../Viz/wilde_sten2.png')
```

```python
color='#DDA0DD'
```

```python
wc = WordCloud(background_color=color, max_words=1000, mask=wilde_mask, stopwords=STOPWORDS)
```

```python
wc.generate_from_text(text)
```

```python
wc.recolor(color_func=grn, random_state=42)
```

```python
plt.imshow(wc)
plt.axis('off')
plt.figure()
```

```python
wc.to_file('../Viz/wilde_viz.png')
```
